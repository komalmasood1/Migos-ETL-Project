```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
spotify_load = "~/Downloads/top2018.csv"
spotify_data = pd.read_csv(spotify_load)
spotify_data_clean = spotify_data.drop(['id'], axis=1)
```

```python
spotify_data_clean.head()
```

```python
billboard_load = "~/Downloads/Hot Stuff.csv"
billboard_data = pd.read_csv(billboard_load)
billboard_data.head()

import datetime 

# billboard_data[billboard_data['Performer']=='Jane Child']
# df[(df['date'] > '2013-01-01') & (df['date'] < '2013-02-01')]
billboard_data['WeekID_DT'] = pd.to_datetime(billboard_data['WeekID'])
# billboard_data.filter(billboard_data['WeekID']=='2/10/1990')]
```

```python
billboard_data = billboard_data[(billboard_data['WeekID_DT'] >= datetime.datetime(year=2018,month=1,day=1)) & \
                     (billboard_data['WeekID_DT'] <= datetime.datetime(year=2018,month=12,day=31))]


billboard_data.head()
```

```python
billboard_cols = ["Song","Performer","Peak Position", "Weeks on Chart"]
billboard_clean = billboard_data[billboard_cols].copy()

billboard_clean = billboard_clean.rename(columns={"Song": "name",
                                                  "Performer": "artists"})


billboard_clean.sort_values(by='Weeks on Chart', ascending=False, inplace=True)
billboard_clean.drop_duplicates("name", inplace=True)

billboard_clean.head()
```

```python
data_merged = pd.merge(spotify_data_clean, billboard_clean, on="name", how="left")
```

```python
data_merged_clean = data_merged.dropna()
data_merged_clean.head()
```

```python
combined_data = data_merged_clean.drop(['artists_y'], axis=1)
combined_data_clean = combined_data.rename(columns={"artists_x" : "artist"})
combined_data_clean.sort_values(by='Weeks on Chart', ascending=False, inplace=True)

combined_data_clean
```
