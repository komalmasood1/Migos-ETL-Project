```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
spotify_data = "Spotify Billboard Correl - Spotify.csv"
```

```python
spotify_df = pd.read_csv(spotify_data)
spotify_df.head()
```

```python
x_energy = spotify_df ["energy"]
y_peak = spotify_df["Peak Position"]
area = np.pi*10
colors = (0,0,0)
```

```python
plt.scatter(x_energy, y_peak, s=area, c=colors, alpha=0.5)
plt.title('Peak vs. Energy')
plt.xlabel('Energy Attribute')
plt.ylabel('Peak Position')
plt.show()
```
